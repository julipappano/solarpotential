# Related Work

Photovoltaic (PV) technology has become a cornerstone of global efforts to decarbonize energy systems. By converting solar radiation directly into electricity, rooftop PV systems offer a decentralized and scalable solution that can be deployed on residential, commercial, and industrial buildings. This makes PV particularly suitable for urban areas with limited land availability. In addition to its technical feasibility, rooftop PV is increasingly valued for its role in supporting energy access, reducing electricity costs, and enhancing grid resilience, especially in regions dependent on imported fossil fuels.

The Caribbean region, with abundant solar resources and a pressing need to reduce reliance on expensive and volatile petroleum imports, is a compelling case for PV deployment. Recent studies have emphasized the potential of rooftop PV to contribute meaningfully to the region's energy transition. For instance, Campo-Ossa et al. (2023) evaluated rooftop PV potential in a low- and moderate-income community in Puerto Rico by estimating potential electricity generation and aligning it with typical household energy needs. Their study highlights the importance of understanding both the physical potential of rooftops and the energy consumption patterns of residents to guide appropriate system design. Variables such as total rooftop area, usable (i.e., suitable) surface, and estimated annual generation were central to their analysis. These same variables are captured in the dataset used in this project, allowing for a similarly grounded evaluation of rooftop solar potential, but applied across different urban contexts in the Dominican Republic. 

Similarly, Oyewo et al. (2025) modeled energy transition scenarios for the Caribbean and projected that solar PV could supply up to 94% of electricity demand by 2050. Although their analysis focuses primarily on utility-scale and offshore solar systems, it also includes distributed generation from rooftop installations. In their model, these are represented as “prosumer systems,” where building occupants both consume and produce electricity using rooftop panels. The study incorporates spatial constraints by modeling scenarios with limited land availability and includes assumptions about how much rooftop area is usable for solar deployment. This reinforces the relevance of building-level variables such as total rooftop surface, suitable area, and installable capacity, as these directly influence how much distributed solar generation can realistically contribute in dense urban areas across the Caribbean.

Within the Dominican Republic specifically, rooftop PV adoption remains modest despite high solar irradiance and growing energy demand. Peña-Bermúdez and Gordillo (2020) reviewed the country’s solar energy policies and found that while Law 57-07 introduced critical incentives for renewable energy investment, photovoltaic systems still account for less than one percent of national electricity production. They attribute this to regulatory instability, inadequate distributed generation planning, and limited access to financing. Severino et al. (2023) further emphasize implementation bottlenecks, noting that while the country has taken steps toward solar development, execution has been slow and fragmented. They identify the need for coordinated, data-driven planning and greater technical capacity to support rooftop PV deployment, particularly in major urban centers such as Santo Domingo and Santiago.

At the global scale, methodological advances have enabled increasingly detailed assessments of rooftop solar potential. Joshi et al. (2021) developed a high-resolution spatiotemporal model to estimate rooftop PV potential globally, using satellite imagery, building footprint data, and machine learning. Their model calculates rooftop availability, solar insolation, and potential electricity generation for individual buildings across more than 130 million square kilometers. Although their work is not focused on prioritizing or classifying buildings, it demonstrates that rooftop-level PV estimates can be scaled using limited field data. The variables used in their model—usable rooftop area, solar resource data, and generation potential—closely correspond to those in this study, reinforcing the relevance of spatially explicit, building-level solar assessments in urban environments such as the Caribbean.


# Data Exploration
### Dataset Overview
The dataset includes detailed rooftop-level data for buildings in 3 cities of the Dominican Republic: Santo Domingo, Santiago de los Caballeros, and San Pedro de Macorís. The file was published in 2024 on the website [Energydata.info](https://energydata.info/dataset/dominican-republic-rooftop-solar-potential-mapping). According to the website, this dataset contains solar rooftop potential data at individual building structure level, which  was gathered by extracting building rooftop footprint polygons from very high-resolution satellite imager. The rooftop angle, obstruction, and shading were taken into account during suitable area calculation. 

As it can be seen below, each data set contains 18 variables and the sum of all observations is 508544. As it shows below, Santo Domingo has a considerably higher number of observation (348,728) when compared to San Pedro de Macorís (145,113) and Santiago de los Caballeros (14,703). 


Each observation in the dataset represents an individual building structure and includes a range of variables describing its physical characteristics and solar potential. The most relevant variables for rooftop PV analysis are:

- **`Surface_area`** (m²): Calculated total rooftop area of the building polygon.

- **`Potential_installable_area`** (m²): Estimated suitable rooftop area for solar panel installation.

- **`Peak_installable_capacity`** (kWp): Estimated installable solar PV capacity.

- **`Energy_potential_per_year`** (kWh): Estimated yearly energy generation from the usable area.

- **`Estimated_installation_efficiency`** (kWh/m²): Ratio calculated using yearly energy potential divided by roof surface area.

- **`Estimated_capacity_factor`**: Ratio calculated using yearly energy potential divided by the installable capacity.

- **`Estimated_tilt`** (degrees): Slope angle of the rooftop.

- **`Estimated_building_height`** (m): Rough estimate of how tall the building is.

- **`Assumed_building_type`**: Describes the use type of the building, for example, commercial, industrial, public, or residential.

- **`Roof_quality`**: Indicates the quality or suitability of a roof for solar panel installation. The value can be High, Medium, or Low.

- **`solar_panels`**: Indicates the existence of solar panels. Values can be Yes or No. Empty column means the investigation was not performed.

## Missing Values

The missing values analysis shows a clear pattern across all three datasets. Four variables (`Unit_installation_price`, `Date_added`, `solar_panels`, and `Comment`) are completely empty in every city. These can be ignored or removed, since they do not add any useful information. The variables `Estimated_building_height` and `Estimated_capacity_factor` have some missing values. `Estimated_building_height` is missing in about 9% of buildings in Santo Domingo and smaller amounts in the other two cities. `Estimated_capacity_factor` is missing in about 2 to 5% of rows. 

```python
# Print missing values for each dataset
for city_name, df in [("Santo Domingo", santodomingo), ("Santiago", santiago), ("San Pedro", sanpedro)]:
    print(f"--- {city_name} ---")
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    for column, count in missing.items():
        print(f"{column}: {count}")
    print()
```

In the case of empty columns, they will be dropped for the analysis since they are not relevant for this project. In the case of relevant variables, we follow the framework proposed by Van Buuren (2018), which classifies missingness into three types: Missing Completely at Random (MCAR), Missing at Random (MAR), and Missing Not at Random (MNAR). 

The bar plots show the share of missing values for two variables in each city, namely `Estimated_building_height` and `Estimated_capacity_factor`. Each bar represents the proportion of buildings in that city that do not have a value recorded for that variable. For example, in the first plot, about 8% of buildings in Santo Domingo are missing a value for `Estimated_building_height`, while the share is lower in Santiago and San Pedro. The second plot shows that missing values in `Estimated_capacity_factor` are generally low and similar across all three cities. These plots are useful because they help identify whether the missing values follow a pattern. In this case, the differences between cities suggest that the missing values are not completely random but may depend on where the building is located. This supports the idea that the data is likely MAR, rather than MCAR. Under the MAR assumption, using grouped imputation methods, such as median imputation by city, is appropriate. The imputation is conducted in the block cell that follows the plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Add city labels
santodomingo["city"] = "Santo Domingo"
santiago["city"] = "Santiago"
sanpedro["city"] = "San Pedro"

# Combine all datasets
df_all = pd.concat([santodomingo, santiago, sanpedro], ignore_index=True)

# Create missing value indicators
df_all["missing_building_height"] = df_all["Estimated_building_height"].isnull()
df_all["missing_capacity_factor"] = df_all["Estimated_capacity_factor"].isnull()

# Plot the proportion of missing values by city
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(data=df_all, x="city", y="missing_building_height", estimator=np.mean, ax=axes[0])
axes[0].set_title("Proportion Missing: Estimated Building Height")
axes[0].set_ylabel("Proportion")

sns.barplot(data=df_all, x="city", y="missing_capacity_factor", estimator=np.mean, ax=axes[1])
axes[1].set_title("Proportion Missing: Estimated Capacity Factor")
axes[1].set_ylabel("Proportion")

plt.tight_layout()
plt.show()
```

```python
# Imputation
df_all["Estimated_building_height"] = df_all.groupby("city")["Estimated_building_height"].transform(
    lambda x: x.fillna(x.median())
)

df_all["Estimated_capacity_factor"] = df_all.groupby("city")["Estimated_capacity_factor"].transform(
    lambda x: x.fillna(x.median())
)

#Remove empty and irrelevant columns 
columns_to_drop = ["Unit_installation_price", "Date_added", "solar_panels", "Comment", "City", "missing_building_height", "missing_capacity_factor"]
existing_columns = [col for col in columns_to_drop if col in df_all.columns]
df_all = df_all.drop(columns=existing_columns)

# Check that NAs are taken care of and irrelevant columns removed
print(df_all[["Estimated_building_height", "Estimated_capacity_factor", "Estimated_Installation_Efficiency","Estimated_tilt","Surface_area", "Potential_installable_area", "Peak_installable_capacity", "Energy_potential_per_year", "Assumed_building_type", "aoi", "Roof_quality", "city"]].isnull().sum())
print("\nUpdated DataFrame shape:", df_all.shape)
print("Remaining columns:\n", df_all.columns.tolist())
```

### Outlier Analysis by City and Variable

The table below shows the number and percentage of outliers for each numerical variable, calculated separately for each city using the standard 1.5 × IQR rule. Across most variables, the share of outliers is consistent and modest,  between 5 and 6%. This pattern holds for rooftop size, installable area, installable capacity, and estimated energy output, suggesting that extreme values are a regular feature of the dataset and not limited to one city.

An exception is found in the variable `Estimated_tilt`. In Santo Domingo, outliers represent only 0.11% of buildings, compared to around 4% in San Pedro and Santiago. This suggests that tilt angles in Santo Domingo are more tightly clustered, possibly reflecting more standardized urban construction.

Overall, the outlier rates are not excessive and do not indicate data errors. However, they highlight a small subset of buildings, typically large, steep, or highly productive, that could strongly influence modeling results. These should be retained for now but treated cautiously in any regression or clustering work.

```python
# Define numeric variables to check
numeric_vars = [
    "Surface_area",
    "Potential_installable_area",
    "Peak_installable_capacity",
    "Energy_potential_per_year",
    "Estimated_tilt",
    "Estimated_building_height",
    "Estimated_capacity_factor",
    "Estimated_Installation_Efficiency"
]

# Function to count outliers using 1.5 * IQR rule
def count_outliers_by_city(df, var):
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return ((df[var] < lower) | (df[var] > upper)).sum()

# Create summary table
outlier_summary = pd.DataFrame(index=numeric_vars)

# Loop over cities
for city in df_all["city"].unique():
    df_city = df_all[df_all["city"] == city]
    outlier_summary[city] = [count_outliers_by_city(df_city, var) for var in numeric_vars]

# Add share of total per city
city_counts = df_all["city"].value_counts().to_dict()
for city in df_all["city"].unique():
    outlier_summary[f"{city}_%"] = (outlier_summary[city] / city_counts[city] * 100).round(2)

# Show result
display(outlier_summary)
```

The outlier table quantifies how many observations fall outside the expected range based on the 1.5 × IQR rule, while the boxplots visually show those same outliers as dots beyond the whiskers. 
For example, in Santo Domingo, variables like `Estimated_building_height` and `Energy_potential_per_year` have a visibly dense cluster of outliers, which aligns with their high outlier counts in the table. In contrast, `Estimated_tilt` shows very few outliers in the boxplot for Santo Domingo, matching the low percentage (0.11%) in the table. # Clustering (h62005716)

## Goal
The goal of this clustering analysis is to identify distinct groups of rooftops with similar solar potential characteristics. Clustering provides a way to segment the dataset without predefined labels, helping to uncover natural patterns in the data. The analysis is conducted with the aggregated data and for each city separately.

## Variable Selection
The chosen variables are those related to the rooftop size and output potential, including: 
* `Surface_area`
* `Potential_installable_area`
* `Peak_installable_capacity`,
* `Energy_potential_per_year`.
 
These features are aligned with the methodology proposed by Joshi et al. (2021), who emphasize rooftop availability and generation potential in global PV assessments. Campo-Ossa et al. (2023) also stress the importance of rooftop suitability and output estimation when evaluating solar deployment in Caribbean communities.

## Method Selection
The clustering method used is the k-means algorithm. Additionally, it should be noted that in the dataset used in this project, variables such as `Energy_potential_per_year` have values in the tens of thousands or even millions (kWh), while others like `Surface_area` or `Peak_installable_capacity` are in the range of hundreds or thousands. If data is not scaled, the clustering algorithm will give much more weight to variables with larger numeric ranges, which can distort the results. To address this, all variables are standardized using StandardScaler, which transforms each feature to have a mean of 0 and a standard deviation of 1. This ensures that each variable contributes equally to the distance calculations used in clustering.


## Model 1: Aggregated Data

### Cluster Selection
To determine the number of clusters in the aggregated dataset, both the Elbow Method and the Davies-Bouldin Score were evaluated. The Elbow plot does not display a pronounced inflection point, suggesting that the rooftop variables used in the clustering model form more of a continuous distribution rather than well-separated groups. This indicates that there is no natural number of clusters that clearly minimizes within-cluster variability. Such a continuous pattern may reflect the heterogeneity of rooftop profiles across urban areas, consistent with the observations of Peña and Gordillo (2020) and Severino et al. (2023), who emphasize the importance of local structural and contextual factors in shaping solar deployment potential. These findings support the decision to also conduct city-level clustering separately.

Despite the absence of a clear elbow, a subtle change in the slope is visible between k = 2 and k = 3, suggesting that three clusters may offer a meaningful simplification of the data. The Davies-Bouldin Score supports this interpretation. Although the lowest score occurs at k = 2 and a second dip at k = 8, the difference across values of k is relatively modest, with no single configuration standing out as an optimal choice. Taken together, these metrics suggest that k = 3 offers a balanced trade-off between model simplicity and the ability to distinguish rooftop profiles. For this reason, three clusters were used in the analysis of the aggregated dataset.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import davies_bouldin_score

# Select variables for clustering
cluster_vars = [
    "Surface_area",
    "Potential_installable_area",
    "Peak_installable_capacity",
    "Energy_potential_per_year"
]

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_all[cluster_vars])

# Elbow method to determine optimal number of clusters
inertia = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 4))
sns.lineplot(x=k_range, y=inertia, marker="o")
plt.title("Elbow Method for Optimal Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.xticks(k_range)
plt.tight_layout()
plt.show()

# Davies-Bouldin score 
db_scores = []
K = range(2, 10)

for k in K:
    preds = KMeans(n_clusters=k, n_init=10, random_state=42).fit_predict(X_scaled)
    score = davies_bouldin_score(X_scaled, preds)
    db_scores.append(score)

# Build and plot DB scores
db_scores_frame = pd.DataFrame({"K": list(K), "Davies-Bouldin Score": db_scores})

plt.figure(figsize=(8, 4))
sns.lineplot(data=db_scores_frame, x="K", y="Davies-Bouldin Score", marker="o")
plt.title("Davies-Bouldin Score by Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Davies-Bouldin Score")
plt.tight_layout()
plt.show()
```

### Cluster Profiles

After fitting the k-means model with three clusters to the aggregated dataset, a highly imbalanced distribution of buildings is observed: Cluster 1 contains the vast majority of rooftops (506,535 buildings), while Clusters 0 and 2 contain only 1,901 and 108 rooftops, respectively. Cluster 1 not only dominates in size but also exhibits the lowest average values across all rooftop variables, including surface area, installable area, peak capacity, and annual energy potential. This suggests that the majority of buildings in the dataset are small-scale structures, likely residential, with limited solar viability. In contrast, Clusters 0 and 2 capture less frequent but substantially different rooftop profiles. Cluster 0 includes mid-sized rooftops, with average energy potential exceeding 500,000 kWh per year, potentially reflecting commercial or institutional facilities. Cluster 2, though representing just 108 buildings, stands out with extremely high values across all metrics, including over 11,000 m² of surface area and more than 2.5 million kWh in estimated annual energy output. These buildings likely correspond to large-scale or industrial facilities with exceptional solar generation potential. Overall, the results support the interpretation that rooftops in the aggregated dataset fall along a continuum, with a large base of low-potential structures and a small number of highly strategic outliers.



```python
# Model with K=3
kmeans_final = KMeans(n_clusters=3, random_state=42, n_init=10)
df_all["cluster"] = kmeans_final.fit_predict(X_scaled)

# Summary statistics by cluster
cluster_summary = df_all.groupby("cluster")[[
    "Surface_area",
    "Potential_installable_area",
    "Peak_installable_capacity",
    "Energy_potential_per_year"
]].mean().round(2)

# Count number of buildings in each cluster
cluster_counts = df_all["cluster"].value_counts().sort_index()

# Show results
print("Number of buildings in each cluster:")
print(cluster_counts)

# Styled display of summary table
cluster_summary.style.set_caption("Table 1. Average Rooftop Characteristics by Cluster").set_table_styles([
    {"selector": "caption", "props": [("caption-side", "top"), ("font-weight", "bold")]}
])
```

```python
# Filter data for Santiago
df_santiago = df_all[df_all["city"] == "Santiago"].copy()

# Define clustering variables 
santiago_cluster_vars = [
    "Surface_area",
    "Potential_installable_area",
    "Peak_installable_capacity",
    "Energy_potential_per_year"
]

# Standardize features
santiago_scaler = StandardScaler()
X_santiago_scaled = santiago_scaler.fit_transform(df_santiago[santiago_cluster_vars])

# Elbow method  
santiago_inertia = []
santiago_k_range = range(2, 10)

for k in santiago_k_range:
    santiago_kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    santiago_kmeans.fit(X_santiago_scaled)
    santiago_inertia.append(santiago_kmeans.inertia_)

# Plot Elbow curve  
plt.figure(figsize=(8, 4))
sns.lineplot(x=list(santiago_k_range), y=santiago_inertia, marker="o")
plt.title("Elbow Method for Optimal Number of Clusters (Santiago)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.tight_layout()
plt.show()

# Davies-Bouldin Score for Santiago
santiago_db_scores = []
for k in santiago_k_range:
    santiago_preds = KMeans(n_clusters=k, n_init=10, random_state=42).fit_predict(X_santiago_scaled)
    santiago_score = davies_bouldin_score(X_santiago_scaled, santiago_preds)
    santiago_db_scores.append(santiago_score)

# Build and plot DB score frame
santiago_db_scores_frame = pd.DataFrame({
    "K": list(santiago_k_range),
    "Davies-Bouldin Score": santiago_db_scores
})

plt.figure(figsize=(8, 4))
sns.lineplot(data=santiago_db_scores_frame, x="K", y="Davies-Bouldin Score", marker="o")
plt.title("Davies-Bouldin Score by Number of Clusters (Santiago)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Davies-Bouldin Score")
plt.tight_layout()
plt.show()
```
